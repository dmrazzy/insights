```python
import pandas as pd
from typing import Dict, List
from ossdirectory import fetch_data
from ossdirectory.fetch import OSSDirectory

data: OSSDirectory = fetch_data()
projects: List[dict] = data.projects
collections: List[dict] = data.collections
```

```python
df_projects = pd.DataFrame(projects)
df_projects.tail()
```

```python
df_collections = pd.DataFrame(collections)
df_collections.head()
```

```python
project_list = df_collections.loc[df_collections['name'] == 'op-retrofunding-4', 'projects'].iloc[0]
print(project_list[:10])
```

```python
def unnest_urls(lst):
    if lst and isinstance(lst[0], dict):
        return [d['url'] for d in lst if isinstance(d, dict) and 'url' in d]
    else:
        return [d['url'] for sublist in lst for d in sublist if isinstance(d, dict) and 'url' in d]
```

```python
nested_list = df_projects.loc[df_projects['name'].isin(project_list), 'github'].to_list()
flat_list = sorted(unnest_urls(nested_list))
print(flat_list[:10])
```
