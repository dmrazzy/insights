```python
# https://dune.com/queries/3162658

START_DATE   = '2023-10-23'
END_DATE     = '2023-10-24'
ATTESTER     = '0x357458739f90461b99789350868cd7cf330dd7ee'
EAS_CONTRACT = '0x4200000000000000000000000000000000000021'
RESOLVER     = '0x94a8e9d0adc7e1c4698ca9ab20e77d6741bf9654'

query = f"""

WITH attestations AS (
  SELECT DISTINCT transaction_hash
  FROM int_superchain_traces_txs_joined
  WHERE
    chain = 'BASE'
    AND block_timestamp BETWEEN DATE '{START_DATE}' AND DATE '{END_DATE}'
    AND to_address_tx = '{ATTESTER}'
    AND from_address_trace = '{EAS_CONTRACT}'
)

SELECT DISTINCT
  block_timestamp,
  to_address_trace AS verified_user
FROM int_superchain_traces_txs_joined
WHERE
  chain = 'BASE'
  AND block_timestamp BETWEEN DATE '{START_DATE}' AND DATE '{END_DATE}'
  AND to_address_trace NOT IN ('{ATTESTER}', '{EAS_CONTRACT}', '{RESOLVER}')
  AND transaction_hash IN (SELECT transaction_hash FROM attestations)

"""

df = oso_client.to_pandas(query)
df
```
